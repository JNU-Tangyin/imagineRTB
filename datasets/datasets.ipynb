## preprocess YOYI datasets 

```python
# download yoyi raw datasets and put it in the 'datasets' folder
import pandas as pd
import numpy as np
# train_data = pd.read_csv('train_set')
# test_data = pd.read_csv('test_set')

# Use 10,000 samples as an example, as the original dataset is too large.
# data = pd.read_csv('datasets//test_set').iloc[:10000]
data = pd.read_csv('E://cikm2016_yoyi_dataset//test_set//test_set').iloc[:10000]
```

```python
# converted to dataframe
result = []
for i in range(len(data)):
    line = data.iloc[i][0]
    print(line)
    parts = line.strip().split(':1\t')
    if any('\t' in part for part in parts):
        parts = [item for part in parts for item in part.split('\t')]
    result.append(parts)
# Extract the first four columns as features and labels
# The first column represents "click", the second  represents "bid_price", and the rest represent other features.
df = pd.DataFrame(result).iloc[:,0:4]
df
```

```python
# convert the values to numeric type
for i in range(4): 
    df.iloc[:,i] = pd.to_numeric(df.iloc[:,i])
df
```

```python
# Re-represent the first column by making continuous processing
from sklearn.preprocessing import MinMaxScaler
d = df.copy()
scaler = MinMaxScaler()
d.iloc[:,1] = scaler.fit_transform(d.iloc[:,1].values.reshape(-1, 1))

def sigmoid(data):
    sigmoid_data = 1 / (1 + np.exp(-data))
    return sigmoid_data

d.iloc[:,1] = [sigmoid(v) for v in d.iloc[:,1]]
d.iloc[:,0] = 0.002*d.iloc[:,1]+ 0.998*d.iloc[:,0]
d
```

```python
# Use a regression model to learn the mapping relation from the feature column to the new representation column.
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
X, y = d.iloc[:,2:4], d.iloc[:,0]

scaler = MinMaxScaler()
X.iloc[:,0] = scaler.fit_transform(X.iloc[:,0].values.reshape(-1,1))
X.iloc[:,1] = scaler.fit_transform(X.iloc[:,1].values.reshape(-1,1))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=50)
model = MLPRegressor(hidden_layer_sizes=(100, 10), activation='relu', solver='adam', random_state=50)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
# save the preprocessed dataframe as txt
df.iloc[:,2] = model.predict(scaler.fit_transform(X))
test_df = df.iloc[1:,0:3]
test_df.to_csv('test.theta_yoyi.txt', sep='\t', index=False)
```

## Process ipinyou datasets

The processing steps for ipinyou were based on those of yoyi, with the addition of the first step, and steps two and three being the same as those for yoyi.

(1) Group the datasets by advertiser objects, resulting in nine sub-datasets. Split the datasets into training and testing sets in a 2:1. You can refer to https://github.com/wnzhang/make-ipinyou-data

(2) Retain the original columns 'click' and 'bid_price', re-represent values of 'click' into continuous values and use regression models to learn the mapping relations between 'click' and other features, generating a third column 'ctr' representing click-through rate.

(3) Save the processed data as a .txt file format. 
Partially processed datasets will be provided as examples for this project.
